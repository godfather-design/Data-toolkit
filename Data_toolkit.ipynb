**Question No. 01. What is NumPy, and why is it widely used in Python?**

**Answer.** NumPy is a powerful Python library for numerical computing. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on them. It's widely used because it’s fast, efficient, and forms the foundation for libraries like Pandas, SciPy, and scikit-learn.

**Question No. 02. How does broadcasting work in NumPy?**

**Answer.** Broadcasting allows NumPy to perform element-wise operations on arrays of different shapes. It automatically expands the smaller array’s shape to match the larger one without copying data. This leads to efficient memory usage and faster computations, especially in mathematical and statistical operations.

**Question No. 03. What is Pandas DataFrame?**

**Answer.** A Pandas DataFrame is a two-dimensional, size-mutable, and labeled data structure. It resembles a table or spreadsheet, with rows and columns. Each column can have a different data type, making it ideal for data manipulation, analysis, and cleaning in Python.

**Question No. 04. Explain the use of groupby() method in Pandas.**

**Answer.** The groupby() method in Pandas is used to split data into groups based on a specific column or condition. Once grouped, you can apply functions like sum, mean, or count to each group. It's essential for data aggregation and analysis based on categories.

**Question No. 05. Why is Seaborn preferred for statistical visualizations?**

**Answer.** Seaborn is preferred because it provides a high-level interface for drawing attractive and informative statistical graphics. It integrates well with Pandas and supports complex visualizations like box plots, violin plots, and heatmaps with simple commands, making data exploration easier.

**Question No. 06. What are the differences between NumPy arrays and Python lists?**
**Answer.** NumPy arrays are more efficient for numerical operations. They support element-wise operations, broadcasting, and occupy less memory than Python lists. Unlike lists, arrays require all elements to be of the same data type, making operations faster and more predictable.

**Question No. 07. What is a heatmap, and when should it be used?**

**Answer.** A heatmap is a graphical representation of data where values are represented using color shades. It is commonly used to visualize correlation matrices, missing data patterns, or any tabular data where relationships between variables need to be understood quickly.

**Question No. 08. What does the term "vectorized operation" mean in NumPy?**

**Answer.** Vectorized operations refer to applying operations on entire arrays without writing explicit loops. This is achieved using NumPy's optimized C backend, which makes computations faster, cleaner, and more efficient compared to traditional looping in Python.

**Question No. 09. How does Matplotlib differ from Plotly?**

**Answer.** Matplotlib is a static plotting library ideal for basic plots and customization, while Plotly offers interactive, web-based visualizations with zoom, hover, and click features. Plotly is better for dashboards and dynamic visualizations, while Matplotlib excels in static and publication-ready graphics.

**Question No. 10. What is the significance of hierarchical indexing in Pandas?**
**Answer.** Hierarchical indexing, or MultiIndexing, allows Pandas to handle multiple index levels in rows or columns. This enables complex data structures like panel data and supports advanced slicing, reshaping, and analysis of multi-dimensional datasets in a 2D DataFrame.

**Question No. 11. What is the role of Seaborn's pairplot() function?**

**Answer.** pairplot() in Seaborn plots pairwise relationships between numerical columns in a dataset. It helps visualize distributions and correlations between variables, making it useful for exploratory data analysis and identifying trends or clusters.

**Question No. 12. What is the purpose of the describe() function in Pandas?**

**Answer.** The describe() function provides summary statistics for numerical columns in a DataFrame, including count, mean, standard deviation, min, and max values. It's helpful for getting a quick overview of the dataset and spotting anomalies or patterns.

**Question No. 13. Why is handling missing data important in Pandas?**

**Answer.** Handling missing data is crucial to maintain data integrity and ensure accurate analysis. Missing values can lead to biased results or errors during computation. Pandas provides functions like fillna(), dropna() to clean and manage incomplete data.

**Question No. 14. What are the benefits of using Plotly for data visualization?**

**Answer.** Plotly offers interactive charts, real-time data updates, and integration with web applications. Its user-friendly syntax, support for 3D plots, and customization features make it a powerful tool for creating dynamic and engaging visualizations.

**Question No. 15. How does NumPy handle multidimensional arrays?**

**Answer.** NumPy supports n-dimensional arrays using the ndarray object. These arrays can have any number of axes, and NumPy provides efficient tools for indexing, slicing, reshaping, and performing operations across dimensions, making it suitable for complex scientific computations.

**Question No. 16. What is the role of Bokeh in Data Visualization?**

**Answer.** Bokeh is a Python library for creating interactive and real-time visualizations in web browsers. It supports a variety of plot types and dashboards, making it ideal for data exploration and sharing insights through interactive visualizations.

**Question No. 17. Explain the difference between apply() and map() in Pandas.**

**Answer.** apply() is used to apply a function along an axis (rows or columns) of a DataFrame, while map() is used for element-wise transformations on a Series. apply() is more flexible and powerful, especially for complex row or column-wise operations.

**Question No. 18. What are some advanced features of NumPy?**

**Answer.** Advanced NumPy features include broadcasting, vectorized operations, masked arrays, memory mapping of large arrays, and linear algebra support. It also supports Fourier transforms, random number generation, and interoperability with other libraries like Pandas and SciPy.

**Question. No. 19. How does Pandas simplify time series analysis?**

**Answer.** Pandas simplifies time series analysis by offering datetime indexing, resampling, frequency conversion, and rolling window calculations. It allows easy parsing of date formats and supports operations like shifting, lagging, and time-based filtering for temporal data analysis.

**Question. No. 20. What is the role of a pivot table in Pandas?**

**Answer.** A pivot table in Pandas summarizes data by grouping it on one or more keys and applying aggregation functions like mean or sum. It helps in reshaping and analyzing large datasets efficiently, similar to pivot tables in Excel.

**Question. No. 21. Why is NumPy's array slicing faster than Python's list slicing?**

**Answer.** NumPy arrays are stored in contiguous memory blocks and are implemented in C. This allows fast indexing and slicing without copying data, unlike Python lists, which are more flexible but slower due to their dynamic and object-based nature.

**Question No. 22. What are some common use cases for Seaborn?**

**Answer.** Seaborn is commonly used for creating statistical visualizations like bar plots, box plots, violin plots, heatmaps, and pairplots. It’s ideal for data exploration, visualizing distributions, correlations, and relationships between variables with minimal code.

**Question No. 1. How do you create a 2D NumPy array and calculate the sum of each row?**

```python
import numpy as np

arr = np.array([[14, 26, 36, 45], [54, 85, 56, 34]])
row_sum = np.sum(arr, axis=1)
print("Row-wise Sum:", row_sum)
```

**Question No. 2. Write a Pandas script to find the mean of a speacific column in DataFrame.**

```python
import pandas as pd

data = {'Marks': [88, 92, 79, 85]}
df = pd.DataFrame(data)
mean_marks = df['Marks'].mean()
print("Mean of Marks:", mean_marks)
```

**Question No. 3. Create a scatter plot using Matplotlib.**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

**Question No. 4. How do you caluclate the correlation matrix using Seaborn and visualize it with a heatmap?**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 5, 6, 7],
    'C': [7, 6, 5, 4]
})
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

**Question No. 5. Generate a bar plot using Plotly.**

**Question No. 6. Create a DataFrame and add a new column based on as existing column.**

```python
df = pd.DataFrame({'Marks': [80, 90, 70]})
df['Grade'] = df['Marks'].apply(lambda x: 'A' if x >= 85 else 'B')
print(df)
```

**Question No. 10.  Create a Histogram using Seaborn to visualize a distribution.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 20, 25, 30, 30, 35, 40]
sns.histplot(data, bins=5, kde=True)
plt.title("Distribution Histogram")
plt.show()
```

**Question No. 11. Perform matrix multiplication using NumPy.**

```python
a = np.array([[1, 2, 7], [3, 4, 9], [7, 3, 7]])
b = np.array([[5, 6, 8], [7, 8, 5], [7, 2, 9]])
result = np.dot(a, b)
print("Matrix Multiplication:\n", result)
```

**Question No. 13. Create a 3D scatter plot using Plotly.**

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3, 4],
    'y': [10, 15, 13, 17],
    'z': [5, 6, 4, 8]
})

fig = px.scatter_3d(df, x='x', y='y', z='z', title='3D Scatter Plot')
fig.show()
```
